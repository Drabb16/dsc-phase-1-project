```python
budgets['worldwide_gross'] = budgets['worldwide_gross'].str.replace(',', '')
budgets['worldwide_gross'] = budgets['worldwide_gross'].str.replace('$', '')
budgets['production_budget'] = budgets['production_budget'].str.replace(',', '')
budgets['production_budget'] = budgets['production_budget'].str.replace('$', '')
```

```python
print(budgets['worldwide_gross'])
```

```python
wwg_list = budgets['worldwide_gross'].tolist()
wwg_list = list(map(int, wwg_list))
```

```python
wwg_list
```

```python
pb_list = budgets['production_budget'].tolist()
pb_list = list(map(int, pb_list))
```

```python
pb_list
```

```python
wwg_array = np.array(wwg_list)
pb_array = np.array(pb_list)
```

```python
roi_array = wwg_array / pb_array
```

```python
roi_array = roi_array * 100
```

```python
roi_list = roi_array.round(2).tolist()
roi_list
```

```python
budgets['roi'] = roi_list 
```

```python
genre_ratings_df = pd.merge(basics,
                            ratings,
                            on = ['tconst'],
                           how = 'left')
```

```python
genre_ratings_df.nlargest(100, 'numvotes').sort_values(by=['numvotes'], ascending = False)
```

From the merged data set we can see popularity, based off number of votes and average rating along with the genres of the title. 

```python
genre_ratings_df.nlargest(100, 'numvotes').mean()
```

```python
genre_ratings_df.nlargest(100, 'numvotes').std()
```

```python
genre_ratings_df['averagerating'].mean()
```

Here we can see the average run time among the top 100 most popular movies is 128 minutes, with a standard deviation of 20 minutes. The average rating is 7.67, which is 1.34 points higher than the average.
